```python
import tensorflow as tf
import numpy as np
keras = tf.keras
```

```python
help(tf.r)
```

```python
class FeatureL2Norm(keras.layers.Layer):
    def __init__(self):
        super(FeatureL2Norm, self).__init__()
    def call(self, feature):
        epsilon = 1e-6
        norm = tf.expand_dims(tf.pow(tf.reduce_sum(tf.pow(feature, 2), axis=-1)+epsilon, 0.5), axis=-1)
        return tf.math.divide(feature, norm)
```

```python
feature = np.ones([64, 3, 3, 4], dtype=np.float32)
print(feature)
```

```python
L2Norm = FeatureL2Norm()
```

```python
x = L2Norm(feature)
print(x)
```
